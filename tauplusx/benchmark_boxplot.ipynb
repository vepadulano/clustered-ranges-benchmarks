```python
import pandas as pd
from matplotlib import pyplot
```

```python
master = pd.read_csv("master.csv", names=["Master"])
enric_pr = pd.read_csv("enric_pr.csv", names=["Enric"])
vincenzo_pr = pd.read_csv("vincenzo_pr.csv", names=["Vincenzo"])
onelistcomp = pd.read_csv("onelistcomp.csv", names=["OneListComp"])
```

```python
merged = pd.concat([master,enric_pr,vincenzo_pr,onelistcomp], axis=1)
```

```python
merged.boxplot(figsize=(10,10))
pyplot.ylim(0.02,0.1)
pyplot.ylabel("Time [s]")
pyplot.savefig("benchmark_boxplot.png")
```
